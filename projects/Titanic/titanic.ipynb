```python
import pandas as pd
import numpy as np

train_data = pd.read_csv('data/train.csv')
predict_data = pd.read_csv('data/test.csv')
```

## 数据缺失处理

```python
# 查看数据缺失情况
for i in range(train_data.shape[1]):
    c = train_data[train_data.columns[i]].count()
    d = {train_data.columns[i]:c/train_data.shape[0]}
    print(d)
```

```python
y = train_data['Survived']
```

### 数据删除

```python
x = train_data.drop(['Survived','Cabin','Name','Ticket'],axis=1)  #先不考虑Cabin,Name,Ticket
predict_x = predict_data.drop(['Cabin','Name','Ticket'],axis=1)
```

## 数据变换

```python
x_after_one_hot = pd.get_dummies(x) #sklearn中的one_hot_encoder有何不同
predict_x_after_one_hot = pd.get_dummies(predict_x)
x_after_one_hot,predict_x_after_one_hot = x_after_one_hot.align(
    predict_x_after_one_hot, join='left', axis=1)  
```

## 数据填充

```python
from sklearn.preprocessing import Imputer
im = Imputer(strategy='most_frequent') # 使用众数进行填充
x_after_imputer = im.fit_transform(x_after_one_hot)
predict_x_after_imputer = im.transform(predict_x_after_one_hot)
```

# 数据建模

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = \
    train_test_split(x_after_imputer, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(train_x, train_y)
```

```python
predict_test_y = rfc.predict(test_x)
```

```python
# 统计错误率
s = 0
test_y_arr = np.array(test_y)
for i in range(predict_test_y.shape[0]):
    if predict_test_y[i] != test_y_arr[i]:
        s = s + 1
print('错误率:%.3lf%%'%(s/predict_test_y.shape[0]*100))
```

# 数据预测

```python
predict_y = rfc.predict(predict_x_after_imputer)
```

# 数据导出

```python
my_submission = pd.DataFrame({'PassengerId': predict_data.PassengerId, 
                              'Survived': predict_y})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
